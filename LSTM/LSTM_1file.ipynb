```python
import os
import PyPDF2

def convert_pdf_to_txt(pdf_path, txt_path):
    try:
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ''
            for page in pdf_reader.pages:
                text += page.extract_text()
            
            with open(txt_path, 'w', encoding='utf-8') as txt_file:
                txt_file.write(text)
            print(f"PDF '{pdf_path}' converted to TXT successfully!")
    except Exception as e:
        print(f"Error converting '{pdf_path}':", e)

def batch_convert_pdfs(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for pdf_file in os.listdir(input_folder):
        if pdf_file.endswith('.pdf'):
            pdf_path = os.path.join(input_folder, pdf_file)
            txt_filename = os.path.splitext(pdf_file)[0] + '.txt'
            txt_path = os.path.join(output_folder, txt_filename)
            convert_pdf_to_txt(pdf_path, txt_path)


```

```python
# Replace 'input_folder' with the folder containing your PDF files
input_folder = 'pdf_data'

# Replace 'output_folder' with the desired folder for the output TXT files
output_folder = 'txt_data'

batch_convert_pdfs(input_folder, output_folder)
```

```python
# Open the file
file = open('txt_Data/Understanding-Visual-Studio-Subscriptions-Administration-Guide-for-VLSC.txt', encoding="utf8")

data= ''
for x in file:

    data += x
    # print(x.strip())

```

```python
print(data)
```

```python

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in data.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=668)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model2 = Sequential()
model2.add(Embedding(668, 100, input_length=16))
model2.add(LSTM(200))
# model.add(LSTM(150))
model2.add(Dense(668, activation='softmax'))
```

```python
model2.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model2.summary()
```

```python
model2.fit(x,y,epochs=100)
```

```python
import time
import numpy as np
text = "VLSC"

for i in range(20):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=16, padding='pre')
  # predict
  pos = np.argmax(model2.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      # time.sleep(2)
```
